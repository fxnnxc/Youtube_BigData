```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import time
```

```python
df = pd.read_csv("댓글1291개0806.csv")
df[:30]
```

```python
def trim_date(x):
    return x[:10]+" "+x[11:19]

df['date'] = pd.to_datetime(df['date'].apply(trim_date))
```

```python
def to_tuple(x):
    return time.mktime(x.timetuple())

df['timestamp'] = df['date'].apply(to_tuple) 
df['timestamp'] -= df['timestamp'].min()
df['timestamp'] /=60
```

```python
hist = plt.hist(df['timestamp'], bins=100)
```

```python
df['hour'] = df['date'].apply(lambda x: str(x)[:15])
df
```

```python
temp = df.groupby(['hour']).count()['comment']
temp = temp.sort_values(ascending=False)
temp[:20]
```
